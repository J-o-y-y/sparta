내 엑셀 데이터는 고객에 대한 데이터를 담고 있는 시트 한 개로 구성되어 있어. 엑셀 데이터는 고객의 [name, age, e-mail, login count, purchase count, total purchase amount]으로 각각의 컬럼이 구성되어 있어. 임의의 100명에 대한 데이터를 생성해서 파일로 저장하는 python 코드를 openpyxl 라이브러리를 사용해서 만들어 줄 수 있어?

네, openpyxl 라이브러리와 random 및 faker 라이브러리를 이용하면 임의의 고객 데이터를 생성하여 엑셀 파일에 저장할 수 있습니다. faker 라이브러리는 이름, 이메일 주소 등의 가짜 데이터를 생성하는 데 유용합니다.

다음은 그러한 코드의 예시입니다:

```python
from openpyxl import Workbook
import random
from faker import Faker

# Workbook 생성
wb = Workbook()
ws = wb.active

# 컬럼명 추가
columns = ['name', 'age', 'e-mail', 'login count', 'purchase count', 'total purchase amount']
ws.append(columns)

fake = Faker()

# 랜덤한 고객 데이터 생성
for i in range(100):  # 100명의 고객 데이터를 생성합니다.
    name = fake.name()
    age = random.randint(20, 60)  # 나이는 20세에서 60세 사이로 임의 설정합니다.
    email = fake.email()
    login_count = random.randint(1, 100)  # 로그인 횟수는 1에서 100 사이로 임의 설정합니다.
    purchase_count = random.randint(1, 20)  # 구매 횟수는 1에서 20 사이로 임의 설정합니다.
    total_purchase_amount = random.randint(1000, 100000)  # 총 구매액은 1000에서 100000 사이로 임의 설정합니다.
    
    row = [name, age, email, login_count, purchase_count, total_purchase_amount]
    ws.append(row)

# 엑셀 파일로 저장
wb.save("customer_data.xlsx")
```

위의 코드는 100명의 고객에 대한 가짜 데이터를 생성하고, 그 데이터를 'customer_data.xlsx'라는 엑셀 파일에 저장합니다. 각각의 고객 정보는 이름, 나이, 이메일 주소, 로그인 횟수, 구매 횟수, 총 구매액으로 구성되어 있습니다. 나이, 로그인 횟수, 구매 횟수, 총 구매액은 임의의 값으로 설정되었습니다. 이 값들의 범위는 필요에 따라 조정하실 수 있습니다.

```python
import openpyxl
import random

# 샘플 데이터 생성
names = ['Alice', 'Bob', 'Charlie', 'Dave', 'Eve', 'Frank', 'Grace', 'Harry', 'Ivy', 'Jack']
ages = [random.randint(20, 50) for _ in range(100)]
emails = [f'{random.choice(names).lower()}@example.com' for _ in range(100)]
login_counts = [random.randint(1, 100) for _ in range(100)]
purchase_counts = [random.randint(1, 20) for _ in range(100)]
total_purchase_amounts = [random.randint(10000, 1000000) for _ in range(100)]

# 엑셀 파일 생성
workbook = openpyxl.Workbook()
worksheet = workbook.active

# 엑셀 파일 헤더 생성
worksheet.append(['name', 'age', 'e-mail', 'login count', 'purchase count', 'total purchase amount'])

# 엑셀 파일 데이터 삽입
for i in range(100):
    worksheet.append([random.choice(names), ages[i], emails[i], login_counts[i], purchase_counts[i], total_purchase_amounts[i]])

# 엑셀 파일 저장
workbook.save('customer_data.xlsx')
```

```python
import openpyxl
import random
import string

# 새로운 엑셀 워크북 생성
wb = openpyxl.Workbook()

# 시트 선택
sheet = wb.active

# 첫 번째 행에 컬럼 이름 추가
sheet.append(['name', 'age', 'e-mail', 'login count', 'purchase count', 'total purchase amount'])

# 100명의 데이터 생성
for i in range(100):
    # 이름 생성 (8자리 영문 소문자)
    name = ''.join(random.choices(string.ascii_lowercase, k=8))
    age = random.randint(20, 60)
    # 이메일 주소 생성 (랜덤 이름 + 실제 도메인)
    email_domain = random.choice(['gmail.com', 'naver.com', 'daum.net', 'yahoo.com'])
    email = f"{name}@{email_domain}"
    login_count = random.randint(1, 50)
    purchase_count = random.randint(1, 20)
    total_purchase_amount = random.randint(10000, 1000000)
    data = [name, age, email, login_count, purchase_count, total_purchase_amount]
    # 시트에 데이터 추가
    sheet.append(data)

# 파일 저장
wb.save("customer_data.xlsx")
```
