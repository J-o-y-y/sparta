```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# 현재 연도를 가져옵니다
current_year = datetime.now().year

# 작년도 시작 날짜와 종료 날짜를 설정합니다
start_date = f'{current_year-1}-01-01'  # 작년도 1월 1일
end_date = f'{current_year-1}-12-31'  # 작년도 12월 31일

# Yahoo Finance를 사용하여 한국전력공사의 주식 데이터를 가져옵니다
data = yf.download('015760.KS', start=start_date, end=end_date)

# 주가 데이터를 그래프로 그립니다
data['Close'].plot()
plt.title('Korea Electric Power Corporation Stock Price')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.show()
```
