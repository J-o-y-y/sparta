```python
import FinanceDataReader as fdr

# 특정 시장의 모든 종목 코드를 가져오는 함수
def get_market_codes(market):
    df = fdr.StockListing(market)
    return df['Code'].tolist()

# KOSPI 시장의 모든 종목 코드를 가져옵니다.
kospi_codes = get_market_codes('KOSPI')

# 각 종목의 가격 데이터를 저장할 딕셔너리
market_data = {}

# 각 종목의 가격 데이터를 가져옵니다.
for code in kospi_codes:
    df = fdr.DataReader(code)
    market_data[code] = df

# 결과를 출력합니다.
print(market_data)

```

```python
import yfinance as yf

# yfinance를 사용해 NDIVIA 주식 정보 가져오기
stock = yf.Ticker("NVDA")
stock_info = stock.info

# 조건 1: 1/PER > 미국 AAA회사채 시장수익률
per = stock_info['trailingPE']
bond_yield = 0.0131 # 2021년 1월 미국 AAA회사채 시장수익률
if per != None and 1/per > bond_yield:
    print("조건 1: 만족")
else:
    print("조건 1: 불만족")

# 조건 2: 배당 수익률 > 회사채 시장수익률×0.67
div_yield = stock_info['dividendYield']
if div_yield != None and div_yield > bond_yield*0.67:
    print("조건 2: 만족")
else:
    print("조건 2: 불만족")

# 조건 3: PBR < 0.65
pbr = stock_info['priceToBook']
if pbr != None and pbr < 0.65:
    print("조건 3: 만족")
else:
    print("조건 3: 불만족")

# 조건 4: 0 < 부채비율 <= 150%
debt_ratio = stock_info['debtToEquity']
if debt_ratio != None and 0 < debt_ratio <= 1.5:
    print("조건 4: 만족")
else:
    print("조건 4: 불만족")

# 조건 5: 유동비율 >= 200%
current_ratio = stock_info['currentRatio']
if current_ratio != None and current_ratio >= 2:
    print("조건 5: 만족")
else:
    print("조건 5: 불만족")
```

```python
import requests
from bs4 import BeautifulSoup
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

# 뉴스 검색 키워드
keyword = "samsung electronics"

# 검색 결과 페이지 URL
url = f"https://search.naver.com/search.naver?where=news&sm=tab_jum&query={keyword}"

# 검색 결과 페이지 요청
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# 헤드라인 스크래핑
headlines = []
for headline in soup.select(".news_tit"):
    headlines.append(headline.text)

# VADER 감성분석 수행
analyzer = SentimentIntensityAnalyzer()
for headline in headlines:
    sentiment_score = analyzer.polarity_scores(headline)
    if sentiment_score["compound"] >= 0.05:
        sentiment = "Positive"
    elif sentiment_score["compound"] <= -0.05:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    print(f"Healine: {headline}")
    print(f"Sentiment: {sentiment}")
    print()

       
```

```python
import requests
from bs4 import BeautifulSoup
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')


# 뉴스 검색 키워드
keyword = "Nvidia"

# 검색 결과 페이지 URL
url = f"https://news.google.com/search?q={keyword}&hl=en-US&gl=US&ceid=US%3Aen"

# 검색 결과 페이지 요청
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# 헤드라인 스크래핑
headlines = []
for headline in soup.find_all("a", class_="DY5T1d"):
    headlines.append(headline.text)

# VADER 감성분석 수행
analyzer = SentimentIntensityAnalyzer()
for headline in headlines:
    sentiment_score = analyzer.polarity_scores(headline)
    if sentiment_score["compound"] >= 0.05:
        sentiment = "Positive"
    elif sentiment_score["compound"] <= -0.05:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    print(f"Healine: {headline}")
    print(f"Sentiment: {sentiment}")
    print()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 카카오 재무제표 페이지 URL
url = 'https://finance.naver.com/item/main.nhn?code=035720'

# requests 모듈로 HTML 페이지 요청
html = requests.get(url).text

# BeautifulSoup으로 HTML 페이지 파싱
soup = BeautifulSoup(html, 'html.parser')

# 재무제표 테이블 선택
table = soup.select_one('div.section.cop_analysis div.sub_section')

# pandas의 read_html 함수를 사용하여 테이블을 데이터프레임으로 변환
df = pd.read_html(str(table), header=0)[0]

# 행 인덱스 설정
df = df.set_index('주요재무정보')

# 출력
print(df)
```

```python
import FinanceDataReader as fdr
import matplotlib.pyplot as plt

# 삼성전자 종목 코드
samsung_code = '005930'

# 최근 1년간의 삼성전자 종가 데이터 가져오기
df = fdr.DataReader(samsung_code, start='2022-05-26', end='2023-05-26')

# 20일 이동평균선 계산
df['MA20'] = df['Close'].rolling(window=20).mean()

# 60일 이동평균선 계산
df['MA60'] = df['Close'].rolling(window=60).mean()

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Close'], label='Close')
plt.plot(df.index, df['MA20'], label='MA20')
plt.plot(df.index, df['MA60'], label='MA60')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Samsung Electronics Moving Average')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import FinanceDataReader as fdr
import matplotlib.pyplot as plt

# 삼성전자 종목 코드
samsung_code = '005930'

# 최근 1년간의 삼성전자 종가 데이터 가져오기
df = fdr.DataReader(samsung_code, start='2022-05-26', end='2023-05-26')

# MACD 곡선 계산
df['MACD'] = df['Close'].ewm(span=12).mean() - df['Close'].ewm(span=26).mean()

# signal 곡선 계산
df['signal'] = df['MACD'].ewm(span=9).mean()

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['MACD'], label='MACD')
plt.plot(df.index, df['signal'], label='signal')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('MACD and Signal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import FinanceDataReader as fdr
import datetime
import pandas as pd
import matplotlib.pyplot as plt

end_date = datetime.datetime.now()  # 현재 시간을 구합니다.
start_date = end_date - datetime.timedelta(days=365)  # 1년 전 날짜를 구합니다.

# 카카오 종목코드: 035720
kakao = fdr.DataReader('035720', start_date, end_date)

# 이동평균선 계산
kakao['MA20'] = kakao['Close'].rolling(window=20).mean()
kakao['MA60'] = kakao['Close'].rolling(window=60).mean()

# 그래프 그리기
plt.figure(figsize=(12,6))
plt.plot(kakao.index, kakao['Close'], label='Close')
plt.plot(kakao.index, kakao['MA20'], label='MA20')
plt.plot(kakao.index, kakao['MA60'], label='MA60')
plt.legend(loc='best')
plt.grid()
plt.show()
```
